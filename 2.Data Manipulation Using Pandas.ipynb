```python
df["Check_in_date"] = pd.to_datetime(df["Check_in_date"])
df["Check_out_date"] = pd.to_datetime(df["Check_out_date"])
```

```python
df.head(10)
```

```python
df.tail()
```

Filtering data

```python
# Only suites
df[df["Room_type"] == "Suite"]
```

```python
# High-value bookings
df[df["Total_bill"] > 20000]
```

```python
# Multiple conditions
df[
    (df["Room_type"] == "Deluxe") &
    (df["Stay_duration"] > 3)
]
```

```python
# Top customers/revenue Data
df.sort_values(by="Total_bill", ascending=False)
```

```python
# Statistical analysis
df["Total_bill"].mean() #Calculates the average revenue generated per booking across all hotel stays.
```

```python
df["Total_bill"].max() #Finds the highest single-booking revenue earned by the hotel.
```

```python
df["Total_bill"].min() #Identifies the lowest revenue generated from a single booking.
```

```python
# Revenue by room type
df.groupby("Room_type")["Total_bill"].sum()
```

```python
# Average stay duration
df.groupby("Room_type")["Stay_duration"].mean()
```

```python
# Feature Engineering
df["Price_per_day"] = df["Total_bill"] / df["Stay_duration"]
```

```python
df.to_csv("Hotel_bookings_updated.csv", index=False)
```
