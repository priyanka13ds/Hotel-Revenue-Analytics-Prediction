```python
import pandas as pd
```

```python
df = pd.read_csv("Hotel_bookings_updated.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum() #Missing values
```

```python
df["Check_in_date"] = pd.to_datetime(df["Check_in_date"])
df["Check_out_date"] = pd.to_datetime(df["Check_out_date"])
```

```python
df.info()
```

```python
df = df[df["Stay_duration"] > 0] #stay duration is possible
```

```python
df = df[df["Total_bill"] > 0] #Revenue cannot be zero or negative.
```

```python
df.duplicated().sum() #finding duplicates
```

```python
df = df[df["Check_out_date"] > df["Check_in_date"]] #logical data validation
```

```python
#Recalculate Stay Duration
df["Stay_duration"] = (
    df["Check_out_date"] - df["Check_in_date"]
).dt.days
```

```python
df.describe() #descriptive statistics analysis
```

```python
df.describe(include='all') #descriptive statistics analysis for both numerical and categorical data
```

```python
df.to_csv("Hotel_bookings_cleaned.csv", index=False)
```
